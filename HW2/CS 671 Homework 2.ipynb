**1.1**

Using the definition of convex functions we can write that
for any $0\leq \alpha\leq 1$ and for all $x_{1},x_{2}$:

\begin{equation}
f\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq
\alpha f\left(x_{1}\right)+(1-\alpha)f\left(x_{2}\right),
\end{equation}

\begin{equation}
g\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq
\alpha g\left(x_{1}\right)+(1-\alpha)g\left(x_{2}\right).
\end{equation}

Adding up these equations:

\begin{eqnarray}
h\left(\alpha x_{1}+(1-\alpha)x_{2}\right) =
f\left(\alpha x_{1}+(1-\alpha)x_{2}\right)+ 
g\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq
\\
\leq\alpha f\left(x_{1}\right)+(1-\alpha)f\left(x_{2}\right) +
\alpha g\left(x_{1}\right)+(1-\alpha)g\left(x_{2}\right)=
\\
=\alpha\left[f(x_1) + g(x_1)\right]+(1-\alpha)\left[f(x_2)+g(x_2)\right]=
\alpha h(x_1)+(1-\alpha)h(x_2),
\end{eqnarray}

or

\begin{equation}
h\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq \alpha h(x_1)+(1-\alpha)h(x_2),
\end{equation}

which is the definition of the convex function.

let's prove that $h(x)=\max(f(x),g(x))$, max of 2 convex functions $f(x)$ and $g(x)$ is also convex.

For that we need to prove, that

$$h\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq \alpha h(x_1)+(1-\alpha)h(x_2).$$

Since $h(x)$ is the maximum of 2 functions $f(x)$ and $g(x)$ we can write that
$$f\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq \alpha h(x_1)+(1-\alpha)h(x_2),$$
$$g\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq \alpha h(x_1)+(1-\alpha)h(x_2.)$$

Due to the definition of a convex function:
$$f\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq\alpha f\left(x_{1}\right)+(1-\alpha)f\left(x_{2}\right),$$
$$g\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq\alpha g\left(x_{1}\right)+(1-\alpha)g\left(x_{2}\right),$$

and we can also write $f(x_1)\leq\max(f(x_1),g(x_1))=h(x_1)$ and $f(x_2)\leq\max(f(x_2),g(x_2))=h(x_2).$
Same for $g(x)$: $g(x_1)\leq\max(f(x_1),g(x_1))=h(x_1)$ and $g(x_2)\leq\max(f(x_2),g(x_2))=h(x_2).$
Combining these equations with the definitions above we can write
$$f\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq\alpha f\left(x_{1}\right)+(1-\alpha)f\left(x_{2}\right)\leq \alpha h(x_1)+(1-\alpha)h(x_2),$$
$$g\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq\alpha g\left(x_{1}\right)+(1-\alpha)g\left(x_{2}\right)\leq \alpha h(x_1)+(1-\alpha)h(x_2),$$

Which means that

$$h\left(\alpha x_{1}+(1-\alpha)x_{2}\right)\leq \alpha h(x_1)+(1-\alpha)h(x_2).$$

# **3 Bias and Variance of Random Forest**

**3.1**

We can prove that fact by induction.

The base case is a training dataset consisting of 1 data point. In that case, we all the data correctly
identified by having it in 1 leaf or making 1 split on any feature.

Now let's consider the induction step. We have the tree, which correctly separated $k$ data points.
We need to prove, that it will also correctly classify $k+1$ data points. If the new extra point $x_{n=k+1}$
has the same feature values as any other $x_k$ that this tree correctly classifies it
since all the overlapping samples are consistently labeled (they are lying down in the same leaf).
If the new sample has a unique set of features, at some point along the tree, we can add the split,
which distinguishes this point from all other points at least by the value of this split. In that case,
we can label $x_{n=k+1}$ with the correct label in his own new leaf.

These two steps establish that a 100% training accuracy tree  can be achieved for any number of samples
if there is no depth limit on the trees in the forest .


**3.2**

The variance by definition is $Var(X)=Cov(X,X)$ and so

$$Var(\bar D)=Var(\dfrac{1}{T}\sum_{i=1}^{T}D_i)=$$
$$Cov(\dfrac{1}{T}\sum_{i=1}^{T}D_i,\dfrac{1}{T}\sum_{j=1}^{T}D_j)=$$
$$\dfrac{1}{T^2}Cov(\sum_{i=1}^{T}D_i,\sum_{j=1}^{T}D_j),$$

We can split this data into 2 separate parts: $i=j$ and $i\ne j$. Since $Cov(X,X)=Var(X)$
we can write

$$Cov(\sum_{i=1}^{T}D_i,\sum_{j=1}^{T}D_j)=$$
$$Var(\sum_{i=1}^{T}D_i) + \sum_{i\ne j}^{T}Cov(D_i,D_j),$$

There are $T^2-T$ possible combinations of $i,j$, where $i\ne j$, and so

$$Var(\sum_{i=1}^{T}D_i) + \sum_{i\ne j}^{T}Cov(D_i,D_j)$$
$$=TVar(D_i) + (T^2-T)Cov(D_i,D_j).$$

By definition of pairwise correlation

$$\rho=Corr(X,Y)=\dfrac{Cov(X,Y)}{\sqrt{(Var(X)Var(Y))}},$$

Thus

$$Cov(X,Y)=\rho\sqrt{(Var(X)Var(Y))}.$$

Using this equation for $Cov(D_i,D_j)$ we can get following expression

$$Var(\bar D)=\dfrac{1}{T^2}(TVar(D_i) + (T^2-T)Cov(D_i,D_j)=$$

$$=\dfrac{1}{T^2}(T\sigma^2+(T^2-T)\rho\sqrt{\sigma^2\sigma^2}))$$
$$=\rho\sigma^2+\sigma^2(\dfrac{1}{T}-\dfrac{1}{T}\rho)=\dfrac{1-\rho}{T}\sigma^2+\rho\sigma^2.$$

**3.3**

As $T$ increases the term $\dfrac{1-\rho}{T}\sigma^2$ in the variance decreases. In the limit, with
 $T\to\infty$ we are getting $\dfrac{1-\rho}{T}\sigma^2\to0$. That means that the variance
 is decreasing with the limitation $\rho\sigma$ when $T\to\infty$.

 **3.4**
 
By increasing the randomness in the random forest we can reduce the correlation between the data and thus
 decrease $\rho$ ($\implies$ decrease $Var(\bar D)$). This can be done by decreasing he size $n$
 of the bootstrap samples and by decreasing the number of features we randomly choose. Both this 
 parameters help reducing variance and bias.

Training the linear SVM algorithm with C=1. Showing the decision boundary in 3D and 2D. The support vectors

```python
from sklearn.svm import SVC

def z(x, y, coeff, inter):
    return x * coeff[0][0] + y * coeff[0][1] + inter

def plotting_boundary(clf, X, C='', limit=(-10, 10)):
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    ax2 = fig.add_subplot(1, 2, 2)
    zArray = np.ones(len(clf.support_vectors_))
    # for 3d plots of the support vectors at the correct z coordinates
    if len(zArray)%2 == 0:
        zArray[0:len(zArray) // 2] -= 2
    else:
        zArray[0:len(zArray) // 2 + 1] -= 2
    ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], zArray, marker='o', c='r', s=100)
    ax.scatter(X[:, 0], X[:, 1], y, marker='o')
    xArray = np.linspace(limit[0], limit[1], 100)
    yArray = np.linspace(limit[0], limit[1], 100)
    xx, yy = np.meshgrid(xArray, yArray)
    ax.plot_surface(xx, yy, -z(xx, yy, clf.coef_, clf.intercept_), alpha=0.2)
    ax2.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], marker='o', c='r', s=130)
    ax2.scatter(X[:len(X)//2, 0], X[:len(X)//2, 1])
    ax2.scatter(X[len(X)//2:, 0], X[len(X)//2:, 1], color='g')
    ax2.scatter(X[:len(X)//2, 0], X[:len(X)//2, 1])
    ax2.scatter(X[len(X)//2:, 0], X[len(X)//2:, 1], color='g')
    coef = clf.coef_[0]
    y_array = -coef[0] / coef[1] * xArray - (clf.intercept_[0]) / coef[1]
    ax2.plot(xArray, y_array)
    print(f'Number of support vectors: {len(clf.support_vectors_[:, 0])} for C={C}')
    fig.suptitle(f'C={C}', fontsize=16)
    plt.show()

clf = SVC(C=1, kernel='linear', random_state=0)
clf.fit(X, y)
print('Accuracy: ', clf.score(X, y))
plotting_boundary(clf, X=X, C=1)
```

# **5 Coding for Logistic Regression (Henry)**

**5.1**

For logistic regression the probability can be represented as follow (vectors are omitted for
brevity):

$$p(y_i=1|X_i\theta)=\dfrac{e^{X_i\theta}}{1+e^{X_i\theta}},$$
$$p(y_i=0|X_i\theta)=1-\dfrac{e^{X_i\theta}}{1+e^{X_i\theta}}=\dfrac{1}{1+e^{X_i\theta}}.$$

Negative log-likelihood:

$$J(\theta)=-\dfrac{1}{n}\sum_i\left[y_i\log{p(y_i=1|X_i\theta)}+(1-y_i)\log{p(y_i=0|X_i\theta)}\right]=$$
$$=-\dfrac{1}{n}\sum_i\left[y_i\log{\dfrac{e^{X_i\theta}}{1+e^{X_i\theta}}}+(1-y_i)\log{\dfrac{1}{1+e^{X_i\theta}}}\right]=$$
$$=-\dfrac{1}{n}\sum_i\left[y_i(\log{e^{X_i\theta}}-\log{(1+e^{X_i\theta}}))-(1-y_i)\log{(1+e^{X_i\theta})}\right]=$$
$$=\dfrac{1}{n}\sum_i\left[-y_iX_i\theta+\log{(1+e^{X_i\theta})}\right].$$

Partial derivative of the negative log-likelihood:

$$\dfrac{\partial J(\theta)}{\partial \theta_j}=$$
$$\dfrac{1}{n}\dfrac{1}{\partial \theta_j}$$
$$\sum_i\left[-y_i\sum_jX_{ij}\theta_j+\log{(1+e^{\sum_jX_{ij}\theta_j})}\right]=$$
$$=\dfrac{1}{n}\sum_i\left[-X_{ij}y_i+X_{ij}\dfrac{e^{X_i\theta}}{1+e^{X_i\theta}}\right]=$$
$$\dfrac{1}{n}\sum_i\left[X_{ij}(-y_i+p(y_i=1|X_i\theta))\right]$$

The update rule:

$$\theta_j \to \theta_j - \eta \dfrac{1}{n}\sum_i\left[-X_{ij}y_i+X_{ij}\dfrac{e^{X_i\theta}}{1+e^{X_i\theta}}\right].$$

**5.2**

**5.3**

Implement a sentiment polarity analyzer using binary logistic regression.

The intercept term is added to the model.

```python
from math import exp
def gradientDescent(data, number_features, rate = 0.1, max_time = 5):
    # copying data
    # and adding intercept
    data_intercept = []
    for d in data:
        new_d1 = set(d[1])
        new_d1.add(number_features)
        data_intercept.append([d[0], new_d1])
    n = len(data_intercept)
    theta = [0] * (number_features + 1)
    for t in range(max_time):
        print(f'epoch: {t}')
        # calculating X_i * theta, since it's the same for all features j.
        sumExp_array = []
        for d in data_intercept:
            sumExp = 0
            for x_ij in d[1]:  # intercept is included in every d[1]
                sumExp += theta[x_ij]
            sumExp_array.append(sumExp)
        for j in range(number_features + 1):
            sum = 0
            for i, d in enumerate(data_intercept):
                if j in d[1]:
                    sum -= d[0]
                    sum += (exp(sumExp_array[i])/(1 + exp(sumExp_array[i])))
            theta[j] -= rate / n * sum
    return theta

theta = gradientDescent(dataTrain, len(words_dict), rate = 0.1, max_time=30)
```

Calculating predictions using theta from the gradient descent.

```python
def predict(data, theta):
    prediction = []
    for d in data:
        f = 0
        for words in d[1]:
            f+= theta[words]
        f += theta[-1]  # intercept term
        if f >= 0:
            prediction.append(1)
        else:
            prediction.append(0)
    return prediction
predictionTest = predict(dataTest, theta)
predictionTrain = predict(dataTrain, theta)
```

Testing the accuracy after 30 epochs. The training error is ~4%, the testing error ~16%.
(Accuracy is ~0.96 and ~0.84, respectively).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
data = pd.read_csv('penguins_trunc.csv')
data_train, data_test = train_test_split(data, test_size=0.2, random_state=2)
```

Using GAM (pygam) with only linear terms. Training accuracy ~0.993, test accuracy ~0.986.

```python
import matplotlib.pyplot as plt
from pygam import LogisticGAM, l
import numpy as np
X, y = np.array(data_train.iloc[:, :-1]), np.array(data_train['Species'])
X_test, y_test = np.array(data_test.iloc[:, :-1]), np.array(data_test['Species'])
gam = LogisticGAM(l(0) + l(1) + l(2)).gridsearch(X, y)
titles = ['CulmenLength', 'CulmenDepth', 'FlipperLength']
gam.summary()
```

```python
fig, axs = plt.subplots(1, 3)
fig.set_size_inches(9, 4)
for i, ax in enumerate(axs):
    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, width=.95)
    ax.plot(XX[:, i], pdep, lw=3)
    ax.plot(XX[:, i], confi, c='r', ls='--')
    ax.set_title(titles[i])
    plt.tight_layout()

print(gam.accuracy(X, y))
print(gam.accuracy(X_test, y_test))
```

Using GAM with splines as basic functions (15 for each feature). As a result, functions are
linear and the training accuracy ~0.993,  test accuracy ~0.986 (same as for
just linear classifiers).

```python
from pygam import s
gam = LogisticGAM(s(0, n_splines=15, spline_order=7) + s(1, n_splines=15)
                  + s(2, n_splines=15)).gridsearch(X, y)
```

```python
fig, axs = plt.subplots(1, 3)
fig.set_size_inches(9, 4)
for i, ax in enumerate(axs):
    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, width=.95)

    ax.plot(XX[:, i], pdep, lw=3)
    ax.plot(XX[:, i], confi, c='r', ls='--')
    ax.set_title(titles[i])
    plt.tight_layout()
print(gam.accuracy(X_test, y_test))
print(gam.accuracy(X, y))
```
